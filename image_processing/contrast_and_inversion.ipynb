```python
image_path = './book.jpeg'
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 이미지 읽기
def load_image(image_path):
    img = cv2.imread(image_path)
    if img is None:
        print("이미지를 찾을 수 없습니다.")
        return None
    return img

# 반전 처리 함수
def strong_inversion(image, intensity=1.5):
    # 이미지 반전 (기본)
    inverted_image = cv2.bitwise_not(image)
    
    # 반전 강도 조절: 원본 이미지와 반전 이미지를 혼합
    strong_inverted = cv2.addWeighted(inverted_image, intensity, image, 1 - intensity, 0)
    
    return strong_inverted
```

```python
def show_image_in_notebook(image, title="Image"):
    # BGR을 RGB로 변환
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # 이미지를 출력
    plt.figure(figsize=(6, 6))
    plt.imshow(image_rgb)
    plt.title(title)
    plt.axis('off')  # 축 숨김
    plt.show()
```

```python
image = load_image(image_path)
```

```python
if image is not None:
    # 반전 및 강도 조절
    inverted_img = strong_inversion(image, intensity=2.0)  # 강도를 높이려면 intensity를 1.0보다 크게 설정
    
    # 반전된 이미지를 노트북에 출력
    show_image_in_notebook(inverted_img, title="Strong Inverted Image")
```
